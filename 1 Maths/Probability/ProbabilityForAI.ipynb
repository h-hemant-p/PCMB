* üé≤ robability Rules
    1. **Range Rule**
        * $[
        0 \leq P(E) \leq 1
        ]$

    2. **Complement Rule**
        * $[
        P(E') = 1 - P(E)
        ]$
    3. **Addition Rule**
        * For mutually exclusive events:
            * $[
            P(A \cup B) = P(A) + P(B)
            ]$
        * For non-mutually exclusive:
            * $[
            P(A \cup B) = P(A) + P(B) - P(A \cap B)
            ]$

        * **Multiplication Rule**

            * For independent events:
            * $[
            P(A \cap B) = P(A) \times P(B)
            ]$
            * For dependent events:
            * $[
            P(A \cap B) = P(A) \times P(B|A)
            ]$

* **Conditional Probability**
    * $[
        P(A|B) = \frac{P(A \cap B)}{P(B)}
    ]$
    * Probability of A happening given that B has already occurred.

    * **Example:** If 70% like ML and 40% like both ML and AI,
        * $[
            P(AI|ML) = \frac{P(AI \cap ML)}{P(ML)} = \frac{0.4}{0.7} = 0.57
        ]$

* üìä**Bayes‚Äô Theore**
    * Bayes‚Äô Theorem describes the probability of an event A given that another event B has occurred.
    * Mattlab Event A ke hone ki probability kitni hai agar Event B hua hai to
    * Bayes‚Äô Theorem ek aisi technique hai jo hume yeh batati hai ki agar koi result mil gaya ho, to us result ke base par kisi reason ke sach hone ki probability kitni hai.
    * Matlab ‚Äî pehle hum reason se result tak jaate hain (forward probability), aur Bayes‚Äô Theorem hume result se reason tak lautne mein madad karti hai (reverse probability).
    * $[
        P(A|B) = \frac{P(B|A)P(A)}{P(B)}
    ]$

    * **Used in:**
        * Naive Bayes Classifier
        * Spam filtering
        * Medical diagnosis

    * **Example:** Agar kisi patient ka test positive aaya hai, to Bayes‚Äô Theorem yeh batati hai ki actually patient ko disease hone ki probability kitni hai.
        * $[
            P(Disease \mid Test+) = \frac{P(Test+ \mid Disease) \times P(Disease)}{P(Test+)}
        ]$
