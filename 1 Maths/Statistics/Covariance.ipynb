# Covariance

* Covariance is a statistical measure indicating the extent to which two random variables change together. 
* It shows the direction, but not the strength, of their linear relationship. 
* A positive covariance suggests variables move in the same direction, while a negative value implies an inverse relationship. Zero covariance indicates no linear relationship. 

* Here is the formula for the sample covariance between two variables $X$ and $Y$:

* $$\text{Cov}(X, Y) = \frac{\sum_{i=1}^{n}(X_i - \bar{X})(Y_i - \bar{Y})}{n-1}$$

    * Where:
        *   $X_i$ and $Y_i$ are the individual data points.
        *   $\bar{X}$ and $\bar{Y}$ are the means (averages) of the $X$ and $Y$ variables, respectively.
        *   $n$ is the number of data points (sample size).
        *   $\sum$ is the summation operator.
