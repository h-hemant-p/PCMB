* **Percentile**
    * **P-th percentile** is the value below which **P% of the data** falls.
        * Example: The 25th percentile (P25) is the value below which 25% of the data lies.
        * Median = 50th percentile (P50)
        
        * How to Calculate Percentiles
            * Step 1: Sort the Data : Arrange the data in **ascending order**.
            * Step 2: Use the Percentile Formula
                * There are several methods, a common one is:
                    [
                    \text{Percentile Rank (Position)} = \frac{P}{100} \times (N + 1)
                    ]
                Where:
                * **P** = percentile you want (e.g., 25 for 25th percentile)
                * **N** = total number of data points
            * Step 3: Find the Value
                * If the position is an **integer**, that position is the percentile.
                * If not, **interpolate** between the two nearest values.
